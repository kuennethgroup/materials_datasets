```python
import pandas as pd

df = pd.read_csv("export.csv", index_col=0)
display(df)
df.property.value_counts()
```

```python
from sentence_transformers import SentenceTransformer

# Add fingeprints

from psmiles import PolymerSmiles as PS

df_egc = df[df.property == "Egc"].reset_index(drop=True)


polyBERT = SentenceTransformer("kuelumbus/polyBERT")

can_smiles = []
for sm in df_egc["smiles"]:
    can_smiles.append(PS(sm).canonicalize.psmiles)
embeddings_pb = polyBERT.encode(can_smiles)

embeddings_ci = []
for sm in df_egc["smiles"]:
    embeddings_ci.append(PS(sm).fingerprint("ci"))

df_egc["fingerprint_polyBERT"] = embeddings_pb.tolist()
df_egc["fingerprint_circluar"] = embeddings_ci
df_egc.to_json("export_egc_fps.json", orient="records")
```

```python
df_egc.to_parquet("export_egc_fps.parquet")
```
